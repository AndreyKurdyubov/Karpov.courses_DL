```python
import torch
from torch.nn import Sigmoid
```

```python
sigm = Sigmoid()
```

```python
# проверка как работает сигмоида
t = torch.tensor([-1000, 0, 0.3, 0.58, 1000])
sigm(t)
```

### Задание
Подсчитайте результат нейронной сети, используя функцию активации сигмоида и ее ошибку. Для расчета ошибки возьмите MSE (Mean Squared Error, среднеквадратичная ошибка). В качестве ответа укажите результат нейронной сети и ее ошибку. Значения округлите до 2 знаков после запятой.

Условия:

Даны значения для входного слоя и веса, соответствующие схеме-картинке. В качестве известного значения целевой переменной (y, target) возьмем 1.
<pre>
(I1)  -  (H1) \
      X         (O1)
(I2)  -  (H2) / 
</pre>
Обозначения: I - Input, H – Hidden layer, O - Output

Входы: I1 = 1, I2 = 0.

Веса первого слоя: w1 = 0.3, w2 = 0.58, w3 = -0.15, w4 = 0.21 
Веса второго слоя: w5 = -0.2, w6 = 1.5

```python
I1, I2 = 1, 0
w1, w2 = 0.3, 0.58
w3, w4 = -0.15, 0.21
w5, w6 = -0.2, 1.5
target = 1

hidden = sigm(torch.tensor([I1*w1 + I2*w3, I1*w2 + I2*w4]))  # H1 , H2 
output = sigm((hidden * torch.tensor([w5, w6])).sum())  # O1
print(output)  # выход сети
print((target - output) ** 2)  # MSE
```
